```julia
using Pkg; Pkg.activate("."); 
using SpectrumImage
```

```julia
fluxes = read_spectrum("time_spec/mean_flux_time.csv", ',', comments=true)
Ilimb = read_spectrum("time_spec/mean_intensity_mu0.20929922_time.csv", ',', comments=true)
Icenter = read_spectrum("time_spec/mean_intensity_mu1.0_time.csv", ',', comments=true);
```

```julia
Î» = fluxes[:, 1]
Fs = fluxes[:, 2:end]
Is_limb = Ilimb[:, 2:end]
Is_center = Icenter[:, 2:end]

F_low = min(minimum(Fs),minimum(Is_limb),minimum(Is_center))
F_high = max(maximum(Fs),maximum(Is_limb),maximum(Is_center))
```

```julia
times = parse.(Float64, split(readline("time_spec/mean_flux_time.csv")[3:end], ',')[2:end]);
```

```julia
times_header = ["$(round(h, digits=2)) h" for h in times];
```

```julia
SpectrumImage.spectrum_gif("test_gif.gif", Î», Fs, F_low=F_low, F_high=F_high, fps=7, header=times_header, show_lambda_range=true, Î»_IR=3800);
```
